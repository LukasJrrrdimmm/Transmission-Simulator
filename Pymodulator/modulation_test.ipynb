```python
import numpy as np

def p500Hz_10A():
	freq = 500
	amp = 10
	return [freq, amp]

def p1KHz_20A():
	freq = 1000
	amp = 20
	return [freq, amp]

def sin_carrier(mm10):
	return np.sin(np.randon.random((mm10,))*mm10)

def cos_carrier(mm10):
	return np.cos(np.randon.random((mm10,))*mm10)

def sign_carrier(mm10):
	return np.sign(np.randon.random((mm10,))*mm10)

def digital_carrier(mm10):
	return np.sin(np.randon.bytes((mm10,)))
```

```python
import numpy as np
import multiprocessing as mp

def mqam(signal, fe, amp):#-> numpy_array
	moduled_signal_mqam1 = []
	moduled_signal_mqam2 = []
	j = 0
	for i in signal:
		moduled_signal_mqam1.append(amp*i*np.cos(2*(np.pi/4)*fe*j)
		moduled_signal_mqam2.append(amp*i*np.sin(2*(np.pi/4)*fe*j)
		j+=3
	return [np.array(moduled_signal_mqam1), np.array(moduled_signal_mqam2)]

def mpsk_modulate1(value, fe, amp, mu, dp, j):
	if value <= mu-dp:
		return(amp*value*np.array(np.sin(2*(np.pi/4)*fe*j)))
	elif value <= mu:
		return(amp*value*np.array(np.sin(2*(np.pi/4)*fe*j + np.pi/4)))
	elif value <= mu+dp:
		return(amp*value*np.array(np.sin(2*(np.pi/4)*fe*j + np.pi/2)))
	else:
		return(amp*value*np.array(np.sin(2*(np.pi/4)*fe*j + 3*np.pi/4)))

def mpsk_modulate2(value, fe, amp, mu, dp, j):
	if value <= mu-dp:
		return(amp*value*np.array(np.cos(2*(np.pi/4)*fe*j)))
	elif value <= mu:
		return(amp*value*np.array(np.cos(2*(np.pi/4)*fe*j + np.pi/4)))
	elif value <= mu+dp:
		return(amp*value*np.array(np.cos(2*(np.pi/4)*fe*j + np.pi/2)))
	else:
		return(amp*value*np.array(np.cos(2*(np.pi/4)*fe*j + 3*np.pi/4)))

def mpsk(signal, fe, amp, mu, dp):
	moduled_signal_mpsk1 = []
	moduled_signal_mpsk2 = []
	j = 0
	for value in signal:
		moduled_signal_mpsk1.append(mpsk_modulate1(value, fe, amp, mu, dp, j))
		moduled_signal_mpsk2.append(mpsk_modulate2(value, fe, amp, mu, dp, j))
	return [moduled_signal_mpsk1, moduled_signal_mpsk2]
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


mu = 10.0
dp = 0.15
val = np.random.normal(mu,dp,500)
freq, amp = p500Hz_10A()
signal = np.sin((np.pi/4)*val)
while 1:
	testchar = '0'
	print("write 'a' 'b' or 'q':")
	testchar = str(input())
	if testchar == 'a':
		mdl1, mdl2 = mqam(signal, freq, amp)
		print([type(mdl1), type(mdl2)])
		df_a = {"original": signal, "moduled p1": mdl1, "moduled p2": mdl2}
		sns.set_style('whitegrid')
		print(pd.DataFrame(df_a).head())
		pd.DataFrame(df_a).plot(kind='line', subplots='true')
		plt.show()
	elif testchar == 'b':
		mdl3, mdl4 = mpsk(val, freq, amp, mu, dp)
		print([type(mdl3), type(mdl4)])
		df_b = {"original": signal, "moduled p1": mdl3, "moduled p2": mdl4}
		sns.set_style('whitegrid')
		print(pd.DataFrame(df_b).head())
		pd.DataFrame(df_b).plot(kind='line', subplots='true')
		plt.show()
	elif testchar == 'q':
		break
	else:
		print("Code Error.")
		break
print(type(val))
print(type(signal))
```
